```python
import pandas as pd
import numpy as np 
import os
print(os.getcwd())

sales = pd.read_csv(r'C:\Users\Chaim\Desktop\Master M2\Smart_Sales_Dashboard\Smart_Sales_Dashboard\data\raw\sales.csv') 
customers = pd.read_csv(r'C:\Users\Chaim\Desktop\Master M2\Smart_Sales_Dashboard\Smart_Sales_Dashboard\data\raw\customers.csv')
```

```python
sales.info
customers.info

print("sales missing data:",sales.isnull().sum())
print("customers missing data:",customers.isnull().sum())
```

```python
sales.drop_duplicates()
```

```python
sales['invoice_no']=sales['invoice_no'].astype(str)
sales['price']=sales['price'].astype(float)
```

```python
import pandas as pd
sales['invoice_date'] = pd.to_datetime(sales['invoice_date'], errors='coerce')
sales['quantity'] = pd.to_numeric(sales['quantity'], errors='coerce')
sales['price'] = pd.to_numeric(sales['price'], errors='coerce')

sales['total'] = sales['quantity'] * sales['price']

last_date = sales['invoice_date'].max()

cust_agg = sales.groupby('customer_id').agg({
    'invoice_date': lambda x: (last_date - x.max()).days,
    'invoice_no': 'nunique',
    'total': 'sum'
}).rename(columns={
    'invoice_date':'Recency',
    'invoice_no':'Frequency',
    'total':'Monetary'
}).reset_index()
```

```python
customers["age"]=customers["age"].fillna(0)
print("customers missing data:",customers.isnull().sum())
```
